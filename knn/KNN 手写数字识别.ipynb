# 1. 准备数据

```python
import os 
import pandas as pd
```

```python
"""
函数功能：得到标记好的数据集
"""
def get_train():
    path = 'digit/trainingDigits'
    trainingFileList = os.listdir(path)
    train = pd.DataFrame()
    img = []
    labels = []
    for i in range(len(trainingFileList)):
        filename = trainingFileList[i]
        txt = pd.read_csv(f'digit/trainingDigits/{filename}',header = None)
        num = ''
        for i in range(txt.shape[0]):
             num += txt.iloc[i,:]
        img.append(num[0])
        filelabel = filename.split('_')[0]
        labels.append(filelabel)
    train['img'] = img
    train['labels'] = labels
    return train
```

# 2.分类器代码的构建

```python
import Levenshtein
```

距离计算公式，使用hamming距离：计算两个等长字串之间对应位置上不同字符的个数，要求输入的两个字符串必须长度一致

```python
from Levenshtein import hamming
# Levenshtein.hamming(str1,str2) #汉明距离格式
```

构建分类器

```python
"""
函数功能：k-近邻算法实现手写数字分类
参数说明：
train：训练集
test： 测试集
k：k-近邻参数，即选择距离最小的k个点
返回：预测好分类的测试集
"""
def handwritingClass(train, test, k):
    n = train.shape[0]
    m = test.shape[0]
    result = []
    for i in range(m):
        dist = []
        for j in range(n):
            d = str(hamming(train.iloc[j,0],test.iloc[i,0]))
            dist.append(d)
        dist_l = pd.DataFrame({'dist':dist, 'labels':(train.iloc[:,1])})
        dr = dist_l.sort_values(by = 'dist')[:k]
        re = dr.loc[:, 'labels'].value_counts()
        result.append(re.index[0])
    result = pd.Series(result)
    test['predict'] = result
    acc = (test.iloc[:,-1]) == test.iloc[:,-2].mean()
    print(f'模型预测准确率为{acc}')
    return test
```
